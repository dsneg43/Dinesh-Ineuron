3. Write a fun which will take input as a dict and give me out as a list of all the values 
even in case of 2 level nesting it should work .

```python
def dict_val(dict_el):
    """Return dictionary values"""
    ret_list = []
    if type(dict_el) == dict:
        val = dict_el.values()
        for i in val:
            if type(i) == dict:
                val1 = i.values()
                for i in val1:
                    ret_list.append(i)
            else:
                ret_list.append(i)
    return(ret_list)                 
```

```python
 dict_val({"a":1,"z":{"x":3, "y":6},"b":2})
```

5. write a function whihc will take multiple list as a input and give me concatnation of all the element as 
and output 

```python
def concat_list(l1,l2,l3):
    """Concat all list"""
    if type(l1) == type(l2) == type(l3) == list:
        return(l1+l2+l3)
    else:
        return("inputs are from different category")
```

```python
l1 = [1,3,5]
l2 = ["2",4,6]
l3 = [7,"name",13]

concat_list(l1,l2,l3)
```

6 . write a function which will be able to take a list as an input return an index of each element like a inbuilt index function but even if we have repetative element it should return index 

```python
def indx(l1,val):
    indx = []
    for i in range(len(l1)):
        if l1[i] == val:
            indx.append(i)
    return(indx)        
        
    
```

8  . write a function which will be able to show your system configuration . 

```python
def system_config():
    import platform
    lst = []
    lst.append({"Machine" :platform.machine()})
    lst.append({"Processor":platform.processor()})
    lst.append({"Architecture":platform.architecture()})
    lst.append({"Platform":platform.platform()})
    lst.append({"system":platform.system()})
    return(lst)
```

```python
system_config()
```
